#### Decision Score 

In this part, I create a decision score to find the perfect arrondissement for the restaurant. Before doing this, I normalize data and rename the column.

```python
#Creating a decision alghoritm
from sklearn import preprocessing

x_test = df_cap[["% hotel of total", "Green/Arrondissement", "Price for m²"]].values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x_test)
df_norm = pd.DataFrame(x_scaled)
df_norm["Arrondissement"] = df_cap["Arrondissement"]
df_norm.columns = ['Hotel', 'Green', 'Price', 'Arrondissement']
df_norm.head(19)
```

```python
#Decision alghoritm
df_norm["Decision score"] = 0.5 * df_norm["Hotel"] + 0.3 * df_norm["Green"] + 0.2 * (1-df_norm["Price"])
df_norm.sort_values(by="Decision score", ascending=False)
```

According to the algorithm, that gives 0.5 of the final score to Hotel, 0.3 to Green and 0.2 to the inverse of Price, the best arrondissement is the 8th. Of course, by prioritizing different elements we will have different results. Still, the lead of the 8th arrondissement on the 16th is pretty strong.

#### Clustering with DBSCAN

Let's cluster the arrondissements with DBSCAN. I used standard scaler on df_cap and I transform the data in array with Numpy.

```python
from sklearn.cluster import DBSCAN
import sklearn.utils
from sklearn.preprocessing import StandardScaler

dataSet = df_cap[['Green/Arrondissement','Number of Hotels']]
dataSet.head()
```

```python
dataSet = np.nan_to_num(dataSet)
dataSet = StandardScaler().fit_transform(dataSet)
dataSet
```

```python
db_clu = DBSCAN(eps=1, min_samples=3).fit(dataSet)
core_samples_mask = np.zeros_like(db_clu.labels_, dtype=bool)
core_samples_mask[db_clu.core_sample_indices_] = True
labels = db_clu.labels_
df_norm["Cluster"]=labels
```

Maximum distance to be considered in the same group = 1, minimum number of samples for cluster = 3.

```python
df_norm.head(20)
df_norm.sort_values(by="Cluster",ascending=True)
```

Very interesting! DBSCAN separated the arrondissements in clusters with a lot of hotels and few green areas and in clusters with a lot of green areas and few luxury hotels. But there is one outlier, the 8th arrondissement! The 8th arrondissement has a huge amount of luxury hotels and a good amount of green areas. In addition, the price is less than €11k for square meter. Thus, let's build our restaurant in the 8th arrondissement!

#### Measuring relationship between Price, Green Ratio and Hotel Ratio

Just to make sure that everything works well, I measure the relationship between the 3 main independent variables of the model we used. I will use scatterploxes.

```python
from sklearn import linear_model

y = np.asanyarray(df_cap[["Price for m²"]])
x = np.asanyarray(df_cap[["Number of Hotels"]])
regr = linear_model.LinearRegression()
regr.fit (x, y)

print ('Coefficients: ', regr.coef_)
print ('Intercept: ',regr.intercept_)

plt.scatter(x, y,  color='blue')
plt.plot(x, regr.coef_[0][0]*x + regr.intercept_[0], '-r')
```

```python
# keep only columns that include venue name, and anything that is associated with location
filtered_columns = ['name', 'categories'] + [col for col in df_resto.columns if col.startswith('location.')] + ['id']
dataframe_filtered = df_resto.loc[:, filtered_columns]

# function that extracts the category of the venue
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']

# filter the category for each row
dataframe_filtered['categories'] = dataframe_filtered.apply(get_category_type, axis=1)

# clean column names by keeping only last term
dataframe_filtered.columns = [column.split('.')[-1] for column in dataframe_filtered.columns]

dataframe_filtered.head()
```

```python
stopwords = set(STOPWORDS)
stopwords.add("add")
stopwords.add("water")
stopwords.add("cup")
stopwords.add("subject")
stopwords.add("tsp")
stopwords.add("minute")
stopwords.add("minutes")
stopwords.add("make")
stopwords.add("vegan")
stopwords.add("Keywords")
stopwords.add("recipe")
stopwords.add("ingredients")
stopwords.add("use")
stopwords.add("pan")

vegan2 = WordCloud(
    background_color='white', 
    width=800, 
    height=400,
    max_words=2000,
    stopwords=stopwords
)

# generate the word cloud
vegan2.generate(vegan)
```
